```python
import pandas as pd
```

```python
cases = list(range(1,9))     # cria uma lista de 1 até n+1 do número de casos exportados
points_lengh = 30            # quantidade de pontos que foram exportados em cada caso
```

```python
df = pd.read_csv("data\export-transient.csv")  # importa o arquivo csv gerado no cfd-post
```

```python
df
```

```python
case_number = []             # cria lista vazia
point_number = []            # cria lista vazia
for case in cases:           # percorre a lista com número de casos definido
    i=0 
    while i < points_lengh:           # enquanto i é menor do que o número de possíveil pontos de detecção exportados
        case_number.append('Case_{}'.format(case))    # nomeia os Casos a cada n total de pontos
        point_number.append('Point_{}'.format(i+1))   # nomeia os pontos de 1 a n para cada caso
        i += 1
```

```python
case_number   # para verificar se está correto

```

```python
points = list(set(point_number))    # cria a lista de pontos sem repetição para usar adiante
points                              # para verificar se está correto
```

```python
df['Case_Number'] = case_number     # acrescenta a coluna com o número de cada caso
df['Point_Number'] = point_number   # acrescenta a coluna com o número de cada ponto
df
```

```python
df['Detected'] = 0 # acrescenta uma coluna vazia
df.head(31)     # imprime para conferência 
```

```python
df2 = df.copy()                                     # cria um novo df igual ao anterior
df2.loc[df2.iloc[:, 3] <= 0.002, 'Detected'] = 0    # na coluna Detected escreverá 0 para fração mássica não detectável e 1 para fração mássica detectável
df2.loc[df2.iloc[:, 3] > 0.002, 'Detected'] = 1   
df2
```

```python
df2[['Point_Number', 'Detected']].groupby('Point_Number').agg(lambda x: list(set(x))).Detected.value_counts().reset_index()  
# mostrar a quantidade de pontos que não detectaram nenhum caso, ou seja, tem o index [0]
```

```python
pcd = []                                                    # cria um dicionário com as informações de ponto, caso e detecção.
final_df = pd.DataFrame()                                   # para a geração do df a partir do dicionário.
for i in points:                                            # caminha cada item da lista points (lista com o nome dos n pontos).
    dfa = df2[df2["Point_Number"] == i]                     # cria um df, a partir do df completo, onde Point_Number é um "Point_i").
    dfb = dfa[dfa["Detected"] == 1]                         # cria um df, a partir do df completo, onde só os casos detectados são escritos.
    lista_dentro = list(dfb["Case_Number"])                 # cria uma lista com a informação dos Case_Number apenas que foram detectados.
    item_pcd = {                            
            "Casos identificados": lista_dentro,            # cria o dicionário item_pcd que contém,
            "Quantidade": len(lista_dentro),                # para cada ponto (que teve pelo menos detecção de um caso), a quantidade e identificação dos casos detectados.
            "Point": i                                      
        }
    final_df = final_df.append(pd.DataFrame(item_pcd))      # cria um df final a partir do dicionário item_pcd.

final_df.sort_values(["Quantidade"], ascending=False)       # ordena o df final por ordem decrescente de quantidade de casos que foram detectados.
```
